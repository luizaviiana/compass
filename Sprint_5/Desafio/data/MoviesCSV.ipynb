-  mostrando todos os valores únicos presentes na coluna, para saber como está o de guerra

```python
df['genero'].unique()
```

- Filtrando o DataFrame df para pegar somente as linhas em que a coluna genero contém a palavra War

```python
df_guerra = df[df['genero'].str.contains('War', na=False)]
df_guerra.head()
```

```python
df_guerra = df_guerra[df_guerra['anoLancamento'] >= 1950]
df_guerra.head()
```

```python
df_guerra = df_guerra[
    (df_guerra['notaMedia'] >= 6) &
    (df_guerra['numeroVotos'] >= 100)
]
```
